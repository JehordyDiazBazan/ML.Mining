```python
# ============================================================
# Plan de minado (Perú) con ML TensorFlow + talud 3x3 + NSR dinámico
# Escenarios: porphyry_cu_au | epithermal_au_ag | polymetallic_zn_pb_ag
# ============================================================

import numpy as np
import pandas as pd
import json
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from matplotlib import cm

np.random.seed(42)
tf.random.set_seed(42)

# ----------------------------
# 0) CONFIG PERÚ (EDITA AQUÍ)
# ----------------------------
SCENARIO = "porphyry_cu_au"  # "epithermal_au_ag" | "polymetallic_zn_pb_ag"

# Tamaño de bloque y malla (comunes en Perú; ajusta a tu ingeniería)
dx, dy, dz = 15, 15, 12  # m
nx, ny, nz = 36, 36, 10

# Precios (USD) - EJEMPLO. Cambia a curvas/contratos del proyecto.
prices_usd = {
    "Au_usd_per_g": 75.0,       # USD/gramo Au
    "Ag_usd_per_g": 0.95,       # USD/gramo Ag
    "Cu_usd_per_t": 8500.0,     # USD/t Cu fino
    "Zn_usd_per_t": 2600.0,     # USD/t Zn fino
    "Pb_usd_per_t": 2100.0      # USD/t Pb fino
}

# Tipo de cambio
fx_usd_to_pen = 3.8  # PEN por USD (ejemplo)

# Payabilities (placeholders; revisa tus contratos/penalidades)
pay = {
    "Au": 0.95,
    "Ag": 0.90,
    "Cu": 0.96,
    "Zn": 0.85,
    "Pb": 0.85
}

# TC/RC aproximados por tonelada de mineral (proxy simple para demo).
# En la práctica, TC (USD/t de concentrado) y RC (c/lb metal) → depende de ley y recuperaciones.
# Aquí usamos "tc_rc_per_t_ore_*" para capturar ese impacto de forma simple.
tc_rc = {
    "Cu_tc_rc_per_t_ore": 2.5,   # USD/t mineral (proxy)
    "Zn_tc_rc_per_t_ore": 1.5,
    "Pb_tc_rc_per_t_ore": 1.2,
    "AuAg_tc_rc_per_t_ore": 0.8
}

# Costos (USD/t mineral) — EJEMPLO Perú; sustituir por tus estimaciones/contratos
costs = {
    "mining_cost_per_t": 3.2,    # Minado open pit promedio (depende de SR/FC)
    "proc_cost_per_t":   9.5,    # Proceso (molienda/flotación/CIP, etc.)
    "ga_cost_per_t":     1.5     # G&A
}
# Penalidades geológicas (USD/t * índice)
penalties = {
    "penalty_hardness":  5.0,    # más duro → más energía/desgaste
    "penalty_clay":      3.5,    # arcillas → reagentes/ensucie
    "depth_penalty_per_m": 0.015 # profundidad → acarreo, bancos
}

# Regalía/gravamen (placeholder): función sobre NSR/t o margen (ajústala a tu caso)
def peru_royalty_per_t(nsr_per_t):
    # Simple: 3% sobre max(NSR/t, 0). Cambia a tu fórmula oficial/contractual.
    return 0.03 * max(nsr_per_t, 0.0)

# Planificación
N_PERIODS = 8
EXTRACT_FRACTION = 0.45  # % masa extraíble (demo)

# -------------------------------------------------
# 1) Datos sintéticos geológicos (con sabor Perú)
# -------------------------------------------------
X, Y, Z = np.meshgrid(np.arange(nx), np.arange(ny), np.arange(nz), indexing="ij")
vol_m3 = dx * dy * dz

def gaussian_field(shape, iters=3, seed=0):
    rng = np.random.default_rng(seed)
    arr = rng.normal(0, 1, size=shape)
    for _ in range(iters):
        arr = (
            np.roll(arr, 1, 0) + arr + np.roll(arr, -1, 0) +
            np.roll(arr, 1, 1) + arr + np.roll(arr, -1, 1) +
            np.roll(arr, 1, 2) + arr + np.roll(arr, -1, 2)
        ) / 9.0
    arr = (arr - arr.min()) / (arr.max() - arr.min() + 1e-9)
    return arr

# Campos "exploración": 0-1
alteration = gaussian_field((nx, ny, nz), seed=11)
clay       = gaussian_field((nx, ny, nz), seed=12)
hardness   = gaussian_field((nx, ny, nz), seed=13)
moisture   = gaussian_field((nx, ny, nz), seed=14)

# Distancia a falla (Andes con estructuras longitudinales)
fault_x = nx//2
dist_fault_m = np.abs(X - fault_x) * dx
depth_m = Z * dz

# Densidad (típicos andesitas/dioritas 2.45–2.80)
density = 2.55 + 0.15*alteration - 0.10*clay
tonnage = density * vol_m3

# Patrones de mineralización por escenario
rng = np.random.default_rng(123)

def make_targets():
    if SCENARIO == "porphyry_cu_au":
        f_Cu = gaussian_field((nx, ny, nz), seed=101)
        f_Au = gaussian_field((nx, ny, nz), seed=102)
        f_Ag = gaussian_field((nx, ny, nz), seed=103)
        # Leyes típicas (aprox): Cu 0.2–0.9%, Au 0.1–0.8 g/t, Ag 1–10 g/t
        Cu = (0.15 + 0.85*f_Cu) * (1 - 0.15*clay) * (1 + 0.10*alteration) + rng.normal(0, 0.02, f_Cu.shape)
        Au = (0.10 + 0.70*f_Au) * (1 - 0.20*clay) + rng.normal(0, 0.08, f_Au.shape)
        Ag = (1.0 + 9.0*f_Ag)  * (1 - 0.15*clay) + rng.normal(0, 0.8,  f_Ag.shape)
        # Recuperaciones por flotación/CIP mixto
        Rec_Cu = np.clip(0.86 - 0.10*hardness - 0.06*clay + 0.04*alteration, 0.55, 0.94)
        Rec_Au = np.clip(0.80 - 0.10*clay - 0.05*moisture + 0.05*alteration, 0.50, 0.92)
        Rec_Ag = np.clip(0.75 - 0.08*clay - 0.05*moisture + 0.03*alteration, 0.50, 0.90)
        return Au, Ag, Cu, None, None, Rec_Au, Rec_Ag, Rec_Cu, None, None
    elif SCENARIO == "epithermal_au_ag":
        f_Au = gaussian_field((nx, ny, nz), seed=111)
        f_Ag = gaussian_field((nx, ny, nz), seed=112)
        # Au 0.3–4 g/t, Ag 10–200 g/t; Cu bajo
        Au = (0.3 + 3.7*f_Au) * (1 - 0.25*clay) + rng.normal(0, 0.20, f_Au.shape)
        Ag = (10 + 190*f_Ag)   * (1 - 0.20*clay) + rng.normal(0, 5.0,  f_Ag.shape)
        Cu = np.zeros_like(Au) + rng.normal(0, 0.005, f_Au.shape) # ~0
        Rec_Au = np.clip(0.86 - 0.08*clay - 0.05*moisture + 0.04*alteration, 0.55, 0.95)
        Rec_Ag = np.clip(0.80 - 0.08*clay - 0.05*moisture + 0.03*alteration, 0.50, 0.92)
        Rec_Cu = np.clip(0.55 - 0.05*clay, 0.40, 0.70)
        return Au, Ag, Cu, None, None, Rec_Au, Rec_Ag, Rec_Cu, None, None
    else:  # "polymetallic_zn_pb_ag"
        f_Zn = gaussian_field((nx, ny, nz), seed=121)
        f_Pb = gaussian_field((nx, ny, nz), seed=122)
        f_Ag = gaussian_field((nx, ny, nz), seed=123)
        Zn = (0.2 + 7.8*f_Zn) * (1 - 0.20*clay) * (1 + 0.10*alteration) + rng.normal(0, 0.10, f_Zn.shape)  # %
        Pb = (0.1 + 6.9*f_Pb) * (1 - 0.20*clay) * (1 + 0.08*alteration) + rng.normal(0, 0.08, f_Pb.shape)  # %
        Ag = (5 + 150*f_Ag)   * (1 - 0.20*clay) + rng.normal(0, 4.0, f_Ag.shape)  # g/t
        # Recuperaciones por flotación Pb/Zn + Ag
        Rec_Zn = np.clip(0.84 - 0.10*hardness - 0.06*clay + 0.04*alteration, 0.55, 0.95)
        Rec_Pb = np.clip(0.86 - 0.10*hardness - 0.06*clay + 0.04*alteration, 0.55, 0.95)
        Rec_Ag = np.clip(0.75 - 0.08*clay - 0.05*moisture + 0.03*alteration, 0.50, 0.90)
        return None, Ag, None, Zn, Pb, None, Rec_Ag, None, Rec_Zn, Rec_Pb

Au_true_gpt, Ag_true_gpt, Cu_true_pct, Zn_true_pct, Pb_true_pct, \
Rec_Au_true, Rec_Ag_true, Rec_Cu_true, Rec_Zn_true, Rec_Pb_true = make_targets()

# Features de exploración
feat_cols = ["alteration","clay","hardness","moisture","dist_fault_m","depth_m","density"]
df = pd.DataFrame({
    "alteration": alteration.flatten(),
    "clay": clay.flatten(),
    "hardness": hardness.flatten(),
    "moisture": moisture.flatten(),
    "dist_fault_m": dist_fault_m.flatten(),
    "depth_m": depth_m.flatten(),
    "density": density.flatten(),
    "x": X.flatten(), "y": Y.flatten(), "z": Z.flatten(),
    "tonnage": tonnage.flatten()
})

# Añadir targets presentes según escenario
def add_if(name, val):
    if val is not None:
        df[name] = val.flatten()

add_if("Au_true_gpt", Au_true_gpt)
add_if("Ag_true_gpt", Ag_true_gpt)
add_if("Cu_true_pct", Cu_true_pct)
add_if("Zn_true_pct", Zn_true_pct)
add_if("Pb_true_pct", Pb_true_pct)
add_if("Rec_Au_true", Rec_Au_true)
add_if("Rec_Ag_true", Rec_Ag_true)
add_if("Rec_Cu_true", Rec_Cu_true)
add_if("Rec_Zn_true", Rec_Zn_true)
add_if("Rec_Pb_true", Rec_Pb_true)

# -----------------------------
# 2) Split / Escalado / Modelos
# -----------------------------
# Construimos matriz de regresión con columnas existentes (multi-salida variable)
target_cols = [c for c in [
    "Au_true_gpt","Ag_true_gpt","Cu_true_pct","Zn_true_pct","Pb_true_pct",
    "Rec_Au_true","Rec_Ag_true","Rec_Cu_true","Rec_Zn_true","Rec_Pb_true"
] if c in df.columns]

X_data = df[feat_cols].values
Y_reg  = df[target_cols].values

X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_reg, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)

def build_regression_model(input_dim, output_dim):
    inp = keras.Input(shape=(input_dim,))
    x = layers.Dense(256, activation="relu")(inp)
    x = layers.Dense(128, activation="relu")(x)
    x = layers.Dropout(0.15)(x)
    out = layers.Dense(output_dim, activation="linear")(x)
    model = keras.Model(inp, out)
    model.compile(optimizer=keras.optimizers.Adam(1e-3), loss="mse", metrics=[keras.metrics.MAE])
    return model

reg_model = build_regression_model(X_train_sc.shape[1], Y_train.shape[1])
_ = reg_model.fit(X_train_sc, Y_train, validation_split=0.2, epochs=45, batch_size=512, verbose=0)
reg_eval = reg_model.evaluate(X_test_sc, Y_test, verbose=0)
print(f"[Regresión] Loss(MSE)={reg_eval[0]:.4f} | MAE={reg_eval[1]:.4f}")

# Predicción para todos
X_all_sc = scaler.transform(X_data)
Y_pred_all = reg_model.predict(X_all_sc, verbose=0)

for i, col in enumerate(target_cols):
    df["pred_"+col.replace("_true","")] = Y_pred_all[:, i]

# Clip por tipo
def clip_if(col, lo, hi):
    if col in df.columns:
        df[col] = np.clip(df[col], lo, hi)

clip_if("pred_Au_gpt", 0, None)
clip_if("pred_Ag_gpt", 0, None)
clip_if("pred_Cu_pct", 0, None)
clip_if("pred_Zn_pct", 0, None)
clip_if("pred_Pb_pct", 0, None)
for c in ["pred_Rec_Au","pred_Rec_Ag","pred_Rec_Cu","pred_Rec_Zn","pred_Rec_Pb"]:
    if c in df.columns:
        df[c] = np.clip(df[c], 0.3, 0.98)

# -----------------------------------------
# 3) NSR por tonelada (dinámico Perú)
# -----------------------------------------
def nsr_per_t_row(row):
    valor_metal_t = 0.0
    # Au/Ag en g/t
    if "pred_Au_gpt" in row and "pred_Rec_Au" in row:
        valor_metal_t += row["pred_Au_gpt"] * row["pred_Rec_Au"] * prices_usd["Au_usd_per_g"] * pay["Au"]
    if "pred_Ag_gpt" in row and "pred_Rec_Ag" in row:
        valor_metal_t += row["pred_Ag_gpt"] * row["pred_Rec_Ag"] * prices_usd["Ag_usd_per_g"] * pay["Ag"]
    # Cu/Zn/Pb en %
    if "pred_Cu_pct" in row and "pred_Rec_Cu" in row:
        valor_metal_t += (row["pred_Cu_pct"]/100.0) * row["pred_Rec_Cu"] * prices_usd["Cu_usd_per_t"] * pay["Cu"]
    if "pred_Zn_pct" in row and "pred_Rec_Zn" in row:
        valor_metal_t += (row["pred_Zn_pct"]/100.0) * row["pred_Rec_Zn"] * prices_usd["Zn_usd_per_t"] * pay["Zn"]
    if "pred_Pb_pct" in row and "pred_Rec_Pb" in row:
        valor_metal_t += (row["pred_Pb_pct"]/100.0) * row["pred_Rec_Pb"] * prices_usd["Pb_usd_per_t"] * pay["Pb"]

    # Costos base + penalidades geológicas
    costos_t = (costs["mining_cost_per_t"] + costs["proc_cost_per_t"] + costs["ga_cost_per_t"]
                + penalties["penalty_hardness"]*row["hardness"]
                + penalties["penalty_clay"]*row["clay"]
                + penalties["depth_penalty_per_m"]*row["depth_m"])

    # TC/RC proxy según escenario
    tc_rc_t = 0.0
    if SCENARIO == "porphyry_cu_au":
        tc_rc_t += tc_rc["Cu_tc_rc_per_t_ore"] + tc_rc["AuAg_tc_rc_per_t_ore"]
    elif SCENARIO == "epithermal_au_ag":
        tc_rc_t += tc_rc["AuAg_tc_rc_per_t_ore"]
    else:  # polymetallic_zn_pb_ag
        tc_rc_t += tc_rc["Zn_tc_rc_per_t_ore"] + tc_rc["Pb_tc_rc_per_t_ore"]

    nsr_before_royalty = valor_metal_t - costos_t - tc_rc_t
    royalty_t = peru_royalty_per_t(nsr_before_royalty)
    return nsr_before_royalty - royalty_t

df["NSR_per_t_usd"] = df.apply(nsr_per_t_row, axis=1)
df["NSR_per_t_pen"] = df["NSR_per_t_usd"] * fx_usd_to_pen
df["NSR_usd"] = df["NSR_per_t_usd"] * df["tonnage"]
df["NSR_pen"] = df["NSR_usd"] * fx_usd_to_pen

# Clasificación dinámica Perú: mineral si NSR/t >= 0
df["is_ore_dyn"] = (df["NSR_per_t_usd"] >= 0.0).astype(int)

# Clasificador (opcional) para frontera bajo estos precios/costos
Xc_train, Xc_test, yc_train, yc_test = train_test_split(df[feat_cols].values, df["is_ore_dyn"].values, test_size=0.2, random_state=42)
scaler_clf = StandardScaler()
Xc_train_sc = scaler_clf.fit_transform(Xc_train)
Xc_test_sc  = scaler_clf.transform(Xc_test)

def build_clf(input_dim):
    inp = keras.Input(shape=(input_dim,))
    x = layers.Dense(256, activation="relu")(inp)
    x = layers.Dense(128, activation="relu")(x)
    x = layers.Dropout(0.2)(x)
    out = layers.Dense(1, activation="sigmoid")(x)
    m = keras.Model(inp, out)
    m.compile(optimizer=keras.optimizers.Adam(1e-3), loss="binary_crossentropy",
              metrics=[keras.metrics.AUC(name="AUC"), keras.metrics.BinaryAccuracy(name="Acc")])
    return m

clf_model = build_clf(Xc_train_sc.shape[1])
_ = clf_model.fit(Xc_train_sc, yc_train, validation_split=0.2, epochs=30, batch_size=512, verbose=0)
clf_eval = clf_model.evaluate(Xc_test_sc, yc_test, verbose=0)
print(f"[Clasificador Perú] Loss={clf_eval[0]:.4f} | AUC={clf_eval[1]:.3f} | Acc={clf_eval[2]:.3f}")

df["pred_is_ore_prob"] = clf_model.predict(scaler_clf.transform(df[feat_cols].values), verbose=0).flatten()

# ---------------------------------------------------------
# 4) Planificación con talud 3x3 y capacidad
# ---------------------------------------------------------
total_extractable_mass = df["tonnage"].sum() * EXTRACT_FRACTION
cap_per_period = total_extractable_mass / N_PERIODS

df["period"] = -1
assigned = np.zeros((nx, ny, nz), dtype=bool)

def can_mine_3x3(row):
    x, y, z = int(row.x), int(row.y), int(row.z)
    if z == nz - 1:  # tope
        return True
    zp = z + 1
    for i in (-1, 0, 1):
        for j in (-1, 0, 1):
            xp, yp = x + i, y + j
            if xp < 0 or xp >= nx or yp < 0 or yp >= ny:
                continue
            if not assigned[xp, yp, zp]:
                return False
    return True

# Prioridad: NSR_usd ponderado por prob. ser mineral
df["prio"] = df["NSR_usd"] * (0.5 + 0.5*df["pred_is_ore_prob"])
candidates = df.sort_values("prio", ascending=False).index.to_list()

for p in range(N_PERIODS):
    cap_left = cap_per_period
    i = 0
    while i < len(candidates):
        idx = candidates[i]
        row = df.loc[idx]
        if row["period"] != -1 or row["NSR_usd"] <= 0:
            i += 1; continue
        if not can_mine_3x3(row):
            i += 1; continue
        if row["tonnage"] <= cap_left:
            df.at[idx, "period"] = p
            assigned[int(row.x), int(row.y), int(row.z)] = True
            cap_left -= row["tonnage"]
            candidates.pop(i)
        else:
            i += 1

plan = (
    df[df["period"] >= 0]
    .groupby("period")
    .agg(
        tonnage=("tonnage","sum"),
        nsr_usd=("NSR_usd","sum"),
        nsr_pen=("NSR_pen","sum"),
        blocks=("period","count")
    )
    .reset_index()
    .sort_values("period")
)
print("\n=== Resumen del Plan (Perú, talud 3x3) ===")
print(plan)
discount_rate = 0.10
plan["discount_factor"] = 1.0 / (1.0 + discount_rate) ** (plan["period"] + 1)
plan["NPV_usd"] = plan["nsr_usd"] * plan["discount_factor"]
plan["NPV_pen"] = plan["NPV_usd"] * fx_usd_to_pen
print(f"\nNPV total: {plan['NPV_usd'].sum():.2f} USD | {plan['NPV_pen'].sum():.2f} PEN")
print(f"Bloques no asignados: {(df['period']<0).sum()} / {len(df)}")

# --------------------------------
# 5) Export: Parquet y GeoJSON
# --------------------------------
cols_export = ["x","y","z","tonnage","NSR_per_t_usd","NSR_per_t_pen","NSR_usd","NSR_pen","period","pred_is_ore_prob"]
for c in ["pred_Au_gpt","pred_Ag_gpt","pred_Cu_pct","pred_Zn_pct","pred_Pb_pct",
          "pred_Rec_Au","pred_Rec_Ag","pred_Rec_Cu","pred_Rec_Zn","pred_Rec_Pb"]:
    if c in df.columns:
        cols_export.append(c)

df_out = df[cols_export].copy()

# Parquet
try:
    df_out.to_parquet("plan_peru.parquet", index=False)
    print("Parquet guardado: plan_peru.parquet")
except Exception as e:
    print("Parquet no guardado (instala pyarrow/fastparquet). Error:", e)

# GeoJSON (Point)
features_geo = []
for _, r in df.iterrows():
    cx = (r["x"] + 0.5) * dx
    cy = (r["y"] + 0.5) * dy
    cz = (r["z"] + 0.5) * dz
    props = {k: float(r[k]) if k in df_out.columns else None for k in df_out.columns}
    props.update({"cx_m": float(cx), "cy_m": float(cy), "cz_m": float(cz)})
    features_geo.append({
        "type": "Feature",
        "geometry": {"type": "Point", "coordinates": [cx, cy, cz]},
        "properties": props
    })
geojson = {"type": "FeatureCollection", "features": features_geo}
with open("plan_peru.geojson", "w", encoding="utf-8") as f:
    json.dump(geojson, f)
print("GeoJSON guardado: plan_peru.geojson")

# --------------------------------
# 6) Viewer simple
# --------------------------------
def view_level_heatmap(df_full, z_level=0, value="NSR_per_t_usd"):
    level = df_full[df_full["z"] == z_level]
    grid = np.full((nx, ny), np.nan)
    for _, r in level.iterrows():
        grid[int(r["x"]), int(r["y"])] = r[value]
    plt.figure(figsize=(7,6))
    plt.imshow(grid.T, origin="lower", cmap="viridis", aspect="equal")
    plt.colorbar(label=value)
    plt.title(f"{value} - Nivel z={z_level} (Perú)")
    plt.xlabel("x"); plt.ylabel("y")
    plt.tight_layout(); plt.show()

def view_3d_scatter(df_full, period=None, value="NSR_per_t_usd"):
    data = df_full if period is None else df_full[df_full["period"] == period]
    if data.empty:
        print("No hay bloques para visualizar.")
        return
    xs = (data["x"]+0.5)*dx; ys = (data["y"]+0.5)*dy; zs = (data["z"]+0.5)*dz
    vals = data[value]
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111, projection='3d')
    p = ax.scatter(xs, ys, zs, c=vals, cmap=cm.viridis, s=6, alpha=0.85)
    fig.colorbar(p, ax=ax, label=value)
    ax.set_xlabel("X (m)"); ax.set_ylabel("Y (m)"); ax.set_zlabel("Z (m)")
    ax.set_title("Bloques 3D (Perú)")
    plt.tight_layout(); plt.show()

# Ejemplos:
# view_level_heatmap(df, z_level=2, value="NSR_per_t_pen")
# view_3d_scatter(df, period=0, value="NSR_per_t_usd")
# ============================================================
# SECCIÓN DE GRÁFICAS DEL MODELO
# (Pegar al final del script, después de entrenar modelos y crear 'plan')
# ============================================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, confusion_matrix, ConfusionMatrixDisplay

# ---------- 0) Si no guardaste los history, ajusta estas dos líneas de entrenamiento ----------
# En tu entrenamiento pon:
# history_reg = reg_model.fit(...); history_clf = clf_model.fit(...)
# (en lugar de "_ = reg_model.fit(...)" o "_ = clf_model.fit(...)")

# ---------- 1) Curvas de entrenamiento / validación ----------
def plot_history_reg(history):
    if history is None:
        print("No hay history_reg disponible.")
        return
    h = history.history
    plt.figure(figsize=(6,4))
    plt.plot(h.get("loss", []), label="loss")
    plt.plot(h.get("val_loss", []), label="val_loss")
    plt.title("Regresión: Loss")
    plt.xlabel("Época"); plt.ylabel("MSE")
    plt.legend(); plt.tight_layout(); plt.show()

    if "mae" in h or "val_mae" in h:
        plt.figure(figsize=(6,4))
        plt.plot(h.get("mae", []), label="mae")
        plt.plot(h.get("val_mae", []), label="val_mae")
        plt.title("Regresión: MAE")
        plt.xlabel("Época"); plt.ylabel("MAE")
        plt.legend(); plt.tight_layout(); plt.show()

def plot_history_clf(history):
    if history is None:
        print("No hay history_clf disponible.")
        return
    h = history.history
    plt.figure(figsize=(6,4))
    plt.plot(h.get("loss", []), label="loss")
    plt.plot(h.get("val_loss", []), label="val_loss")
    plt.title("Clasificador: Loss")
    plt.xlabel("Época"); plt.ylabel("Loss")
    plt.legend(); plt.tight_layout(); plt.show()

    if "AUC" in h or "auc" in h or "val_auc" in h:
        # Las claves pueden variar de 'auc' a 'AUC' según versión
        auc_key = "auc" if "auc" in h else ("AUC" if "AUC" in h else None)
        val_auc_key = "val_auc" if "val_auc" in h else ("val_AUC" if "val_AUC" in h else None)
        if auc_key:
            plt.figure(figsize=(6,4))
            plt.plot(h.get(auc_key, []), label="AUC")
            if val_auc_key: plt.plot(h.get(val_auc_key, []), label="val_AUC")
            plt.title("Clasificador: AUC")
            plt.xlabel("Época"); plt.ylabel("AUC")
            plt.legend(); plt.tight_layout(); plt.show()

# Llama (si definiste history_reg / history_clf):
try:
    plot_history_reg(history_reg)
except NameError:
    print("history_reg no definido (ajusta tu fit para guardarlo).")
try:
    plot_history_clf(history_clf)
except NameError:
    print("history_clf no definido (ajusta tu fit para guardarlo).")

# ---------- 2) Paridad (y_pred vs y_true) por target ----------
# Usa X_test_sc / Y_test del bloque de regresión principal
try:
    Y_pred_test = reg_model.predict(X_test_sc, verbose=0)
    # Mapeo nombre->índice según target_cols
    target_index = {name:i for i, name in enumerate(target_cols)}
    for name, i in target_index.items():
        y_true = Y_test[:, i]
        y_pred = Y_pred_test[:, i]
        plt.figure(figsize=(5,5))
        plt.scatter(y_true, y_pred, s=10, alpha=0.5)
        mn, mx = np.min([y_true.min(), y_pred.min()]), np.max([y_true.max(), y_pred.max()])
        plt.plot([mn, mx], [mn, mx], linestyle="--")
        plt.title(f"Paridad: {name.replace('_true','')}")
        plt.xlabel("Valor verdadero"); plt.ylabel("Predicción")
        plt.tight_layout(); plt.show()
except Exception as e:
    print("No se pudo graficar paridad de regresión:", e)

# ---------- 3) ROC y matriz de confusión del clasificador dinámico ----------
# Usa Xc_test_sc / yc_test del bloque del clasificador NSR/t
try:
    y_score = clf_model.predict(Xc_test_sc, verbose=0).ravel()
    fpr, tpr, thr = roc_curve(yc_test, y_score)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(6,5))
    plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
    plt.plot([0,1], [0,1], linestyle="--")
    plt.title("ROC - Clasificador (NSR/t ≥ 0)")
    plt.xlabel("FPR"); plt.ylabel("TPR")
    plt.legend(); plt.tight_layout(); plt.show()

    # Matriz de confusión con umbral 0.5 (puedes cambiarlo)
    y_pred_bin = (y_score >= 0.5).astype(int)
    cm = confusion_matrix(yc_test, y_pred_bin, labels=[0,1])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Esteril","Mineral"])
    disp.plot(values_format="d")
    plt.title("Matriz de confusión (umbral=0.5)")
    plt.tight_layout(); plt.show()
except Exception as e:
    print("No se pudo graficar ROC/confusión:", e)

# ---------- 4) Distribuciones de NSR ----------
try:
    # Histograma NSR/t (USD)
    plt.figure(figsize=(6,4))
    plt.hist(df["NSR_per_t_usd"], bins=40)
    plt.title("Distribución NSR por tonelada (USD/t)")
    plt.xlabel("NSR/t (USD)"); plt.ylabel("Frecuencia")
    plt.tight_layout(); plt.show()

    # Histograma NSR por bloque
    plt.figure(figsize=(6,4))
    plt.hist(df["NSR_usd"], bins=40)
    plt.title("Distribución NSR por bloque (USD)")
    plt.xlabel("NSR (USD)"); plt.ylabel("Frecuencia")
    plt.tight_layout(); plt.show()
except Exception as e:
    print("No se pudo graficar distribuciones de NSR:", e)

# ---------- 5) Plan por periodo (toneladas vs NSR) ----------
try:
    # Asegurar que 'plan' exista
    if 'plan' in globals():
        # Barras: toneladas | Línea: NSR (USD)
        plt.figure(figsize=(7,4))
        plt.bar(plan["period"], plan["tonnage"])
        plt.title("Toneladas por periodo")
        plt.xlabel("Periodo"); plt.ylabel("Toneladas")
        plt.tight_layout(); plt.show()

        plt.figure(figsize=(7,4))
        plt.plot(plan["period"], plan["nsr_usd"], marker="o")
        plt.title("NSR por periodo (USD)")
        plt.xlabel("Periodo"); plt.ylabel("NSR (USD)")
        plt.tight_layout(); plt.show()

        # (opcional) NPV por periodo
        if "NPV_usd" in plan.columns:
            plt.figure(figsize=(7,4))
            plt.plot(plan["period"], plan["NPV_usd"], marker="o")
            plt.title("NPV por periodo (USD)")
            plt.xlabel("Periodo"); plt.ylabel("NPV (USD)")
            plt.tight_layout(); plt.show()
    else:
        print("No existe 'plan' en el entorno.")
except Exception as e:
    print("No se pudo graficar plan por periodo:", e)
```
